```python
class Animal:
    def __init__(self, name):
        self.name = name

    def speak(self):
        raise NotImplementedError("Subclass must implement this method")


class Dog(Animal):
    def speak(self):
        return f"{self.name} says woof!"


my_dog = Dog("Buddy")
print(my_dog.speak())
```

```python
def find_largest(a, b, c):
    return max(a, b, c)

print(find_largest(5, 10, 22))
```

```python
def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n - 1)

print(factorial(5))
```

```python
def add(a, b):
    return a + b

def subtract(a, b):
    return a - b

def multiply(a, b):
    return a * b

def divide(a, b):
    return a / b if b != 0 else "Division by zero error"
```

```python
class Rectangle:
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width

    def perimeter(self):
        return 2 * (self.length + self.width)

rect = Rectangle(10, 5)
print(rect.area())
print(rect.perimeter())
```

```python
counter = 1

def increment_counter():
    global counter
    counter += 1
    print("Inside function, counter =", counter)

print("Before function call, counter =", counter)
increment_counter()
print("After function call, counter =", counter)
```
