```python
import numpy as np
# 1D array
array1d = np.array([1, 2, 3])
print(array1d) # Output: [1 2 3]
# 2D array
array2d = np.array([[1, 2], [3, 4]])
print(array2d)
```

```python
import numpy as np

# 1D array input
array1d = np.array(list(map(int, input("Enter 1D array elements, separated by space: ").split())))

# 2D array input
rows, cols = map(int, input("Enter number of rows and columns for the 2D array, separated by space: ").split())
array2d = np.array([list(map(int, input(f"Enter row {i+1}: ").split())) for i in range(rows)])

# Display the arrays
print("1D array:", array1d)
print("2D array:\n", array2d)
```
