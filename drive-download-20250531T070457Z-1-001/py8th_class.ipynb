```python
class Student:
    def __init__(self, name, student_id, department):
        self.name = name
        self.student_id = student_id
        self.department = department

    def display_info(self):
        print(f"Name: {self.name}, ID: {self.student_id}, Department: {self.department}")

def main():
    students = []

    # Input details for 12 students
    for i in range(12):
        print(f"\nEnter details for student {i + 1}:")
        name = input("Enter name: ")
        student_id = input("Enter ID: ")
        department = input("Enter department: ")

        # Create a new Student object and append it to the list
        student = Student(name, student_id, department)
        students.append(student)

    # Display all student information
    print("\nStudent Information:")
    for student in students:
        student.display_info()

if __name__ == "__main__":
    main()

```

```python
class Rectangle:
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        # Calculates and returns the area of the rectangle
        return self.length * self.width

    def perimeter(self):
        # Calculates and returns the perimeter of the rectangle
        return 2 * (self.length + self.width)

# Test case
rect = Rectangle(10, 5)
print(rect.area())      # Output: 50
print(rect.perimeter()) # Output: 30
```

```python
class BankAccount:
    def __init__(self, account_number, balance=0):
        self.account_number = account_number
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount

    def withdraw(self, amount):
        if self.balance >= amount:
            self.balance -= amount
        else:
            return "Insufficient funds"

    def get_balance(self):
        return self.balance

# Test case
account = BankAccount(101, 1000)
account.deposit(500)
print(account.get_balance())  # Output: 1500

result = account.withdraw(200)
if result != "Insufficient funds":
    print(account.get_balance())  # Output: 1300
else:
    print(result)  # If insufficient funds, it will print the message
```

```python
class BankAccount:
    def __init__(self, account_number, balance=0):
        self.account_number = account_number
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount

    def withdraw(self, amount):
        if self.balance >= amount:
            self.balance -= amount
        else:
            return "Insufficient funds"

    def get_balance(self):
        return self.balance


class Bank:
    def __init__(self):
        self.accounts = []

    def add_account(self, account):
        # Add an account object to the bank's list of accounts
        self.accounts.append(account)

    def total_funds(self):
        # Calculate the total funds in all bank accounts
        return sum(account.get_balance() for account in self.accounts)


# Test case
account1 = BankAccount(101, 1000)
account2 = BankAccount(102, 1500)

# Create a bank and add accounts to it
bank = Bank()
bank.add_account(account1)
bank.add_account(account2)

# Print the total funds in the bank
print(bank.total_funds())  # Output: 2500
```

```python
class Rectangle:
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        # Area of rectangle = length * width
        return self.length * self.width

    def perimeter(self):
        # Perimeter of rectangle = 2 * (length + width)
        return 2 * (self.length + self.width)

# Example usage:
rect = Rectangle(5, 3)
print(rect.area())         # Output: 15
print(rect.perimeter())    # Output: 16
```
