```python
import numpy as np
data = np.random.randn(10000)
mean_val = np.mean(data)
median_val = np.median(data)
std_dev = np.std(data)

print(f"Mean: {mean_val}, Median: {median_val}, Std Dev: {std_dev}")

min_val = np.min(data)
max_val = np.max(data)

print(f"Min: {min_val}, Max: {max_val}")

greater_than_mean = np.sum(data > mean_val)

print(f"Values greater than mean: {greater_than_mean}")

normalized_data = (data - min_val) / (max_val - min_val)

print("Normalized Data (first 10 values):", normalized_data[:10])








```
